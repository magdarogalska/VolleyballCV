```python
from ultralytics import YOLO
model_path ='/content/runs/detect/train/weights/best.pt'
# Load a model
model = YOLO(model_path)  # load a custom model

threshold = 0.5

while ret:

    volley_frames = model.predict(frame)
    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result
        cx = int((x1 + x2)/2)
        cy = int((y1 + y2)/2)
        if score > threshold:
            cv2.imwrite(f'/content/frame{frameNum}.PNG',frame) #write the frame to an image
            output_dir = f'/Users/infrered/Desktop/VolleyBall_Frames/Test_Frames_from_Video/images/frame{frameNum}.txt'
            cx1 = cx / 1920
            cy1 = cy / 1080
            str_c1 = str(cx1)
            str_c2 = str(cy1)
            coord = str_c1 + "," + str_c2
            frame_Num = str(frameNum)
            combined_string = frame_Num + "," + coord
            print("Combined String: ", combined_string)
            with open(output_dir, 'w') as file:
                    file.write(combined_string)
            anno_frame = frame.copy()
          cv2.circle(anno_frame, (cx,cy), 15, (255,0,0), 4)
          cv2.imwrite(f'/Users/infrered/Desktop/VolleyBall_Frames/Test_Frames_from_Video/images/frame{frameNum}_annotated.PNG',anno_frame)
          cv2.circle(frame, (cx, cy),20,(255, 0, 0), -1)
          cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
          cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
