```python
import os
f = open("tracknet_labels.csv", "a")



for i in range(1440):
  try:
    f2 = open(f"/content/drive/MyDrive/VolleyBall_model/train/labels/frame_C{i}.txt", "r")
    first = f2.read(1)
    if not first:
        continue
    else:
        f2.seek(0)
        line = f2.readline()
        ls = line.split(' ')
        f.write(f"\nframe_C{i}, {ls[1]}, {ls[2]}")
  except OSError:
    print("Could not open/read file")
    continue


f.close()
```
